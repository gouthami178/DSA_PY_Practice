```python
### Count num of ways to reach upstairs ###
### Can take either 1 or 2 steps at a time ###
def CountNumOfWays(n):
    if n <= 1:
        return n
    else:
        return CountNumOfWays(n-1) + CountNumOfWays(n-2)


##Driver code
n = 6 #num steps
res = CountNumOfWays(n)
print("number of ways to reach",n,"steps is",res, "ways")
```
