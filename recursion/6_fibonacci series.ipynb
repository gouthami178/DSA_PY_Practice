```python
### fibonacci series method 1
def fibonacciSeries(t1,t2,n):
    if n > 0 :
        print(t1+t2)
        return fibonacciSeries(t2,t1+t2,n-1)
    
## Driver Code
# n elements to print
n = 8
t1 = 0
t2 = 1
print(t1)
print(t2)
fibonacciSeries(t1,t2,n-2)
```

```python
### fibonacci series method 2
def fibonacciSeries(n):
    if n <= 1 :
        return n
    else:
        return fibonacciSeries(n-1) + fibonacciSeries(n-2)
    
## Driver Code
# n elements to print
n = 8
for i in range(n):
    print(fibonacciSeries(i))
```
