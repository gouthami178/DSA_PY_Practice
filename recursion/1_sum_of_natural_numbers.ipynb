```python
###Sum of natural numbers - using recursion ###
def SumNumbers(num):
    if num == 0:
        return 0 

    return num + SumNumbers(num-1)

    
#input num for the sum to be found
num = 7
res = SumNumbers(num)
print("Sum of top",num, "is", res) 
```

```python
###Sum of natural numbers###
def SumNumbers(num):
    total = 0
    for i in range(1,num):
        total = total + i
    return total

    
#input num for the sum to be found
num = 5
res = SumNumbers(num)
print("Sum of top",num, "is", res) 
```
