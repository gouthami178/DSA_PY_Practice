```python
### Implementation of Binary Search (without recursion) ###
### input is sorted array                               ###
def BinarySearch(in_arr, ele, start_idx, end_idx):
    while start_idx<=end_idx:
        mid = start_idx + (end_idx - start_idx)//2
        if(in_arr[mid] == ele):
            return mid
        elif in_arr[mid] > ele: #ele is in first half
            end_idx = mid - 1
        else:   #ele is in 2nd half of array
            start_idx = mid + 1
    return -1

# Driver Code
in_arr = [2, 5, 10, 14, 18, 22, 27, 35, 40, 59]
ele = 27
start_idx = 0
end_idx = len(in_arr) - 1

# Function call
res = BinarySearch(in_arr, ele, start_idx, end_idx)
print("search element is present at index",res)
```
