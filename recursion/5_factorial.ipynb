```python
### Factorial with recursion ###
def FindFactorial(n):
    result = 1
    if (n == 0 or n == 1):
        return n
    else:
        return n*FindFactorial(n-1)

## Driver Code
# factorial for the number to find
n = 8
res = FindFactorial(n)
print("factorial of",n,"is = ", res)
```

```python
### Factorial without recursion ###
def FindFactorial(n):
    result = 1
    if (n == 0 or n == 1):
        return n
    else:
        for i in range(2,n+1):
            result = result*i
        return result

## Driver Code
# factorial for the number to find
n = 8
res = FindFactorial(n)
print("factorial of",n,"is = ", res)
```
