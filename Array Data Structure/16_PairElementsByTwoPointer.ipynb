```python
### find a pair such that sum of pair elements = target value
## Given sorted array
def findPair(arr, target):
    l = 0
    r = len(arr) - 1
    i =0

    for i in range(len(arr)-1):
        if arr[l] + arr[r] == target :
            return l,r
        elif arr[l] + arr[r] > target :
            r = r-1
        else:
            l = l+1
    return -1,-1

## Driver code
arr = [20, 40, 50, 75, 120, 145, 200]
target = 60
## function calling
result = findPair(arr, target)
print(result)
```
