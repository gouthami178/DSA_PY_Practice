```python
### Given array give most frequency element value.
### if it doesnt exisit print -1###
## Approach 3 - Boyer Moore Voting Algorithm
def majorityElement(arr):
    candidate = -1
    count = 0
    n = len(arr)
    for i in range(n):
        if(count == 0):
            candidate = arr[i]

        if candidate == arr[i]:
            count = count + 1
        else:
            count = count -1

     #MAjority element check
    cnt = 0
    for i in range(n):
        if(candidate == arr[i]):
            cnt += 1
        if(cnt > n/2):
            print("majority element is",candidate)
            break
    
    if(cnt <= n/2): 
        print("there is no majority element") 

## Driver code
arr = [2, 2, 1, 1, 1, 2, 2]
#arr = [2, 2, 7, 3, 4]
majorityElement(arr)
```
