```python
''' Given tree colors Red, Green, Blue each has values 0,1,2 respectively
input: these three colors are in an array undorted
output: sort these colors
input: [ 1,1,0,2,0,2]
output: [0,0,1,1,2,2]        '''
def sortColors(nums):
    left = curr = 0
    right = len(nums)-1

    while(left <= right):
        if nums[curr] == 0: #swap to left side
            nums[left], nums[curr] = nums[curr], nums[left]
            left+=1
            curr+=1
        elif nums[curr] == 2: #swap to right side
            nums[right], nums[curr] = nums[curr], nums[right]
            right -=1
        else:
            curr +=1
    return nums

## Driver code
nums = [0, 2, 1, 1, 0, 2]
result = sortColors(nums)
print(result)
```
