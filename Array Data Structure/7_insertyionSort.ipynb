```python
### Insertion Sort ###
def insertionSort(arr):
    n = len(arr)
    for i in range(1,n):
        new_ID = i
        for j in range(0,i):
            #print(new_ID ,j,arr[j],arr[new_ID])
            if arr[new_ID] < arr[j]:
                arr[j],arr[new_ID] = arr[new_ID],arr[j]
            
    return arr

arr = [9, 5, 1, 4, 3]
## function calling
result = insertionSort(arr)
print("Sorted array after applying the insertion sort is", result)
```
