```python
## Using Divide and Conquer Approach find min and max in an array
## Time Complexity : O(n)
def findMaxandMin(arr, i, j):
    if i==j:  #single element min,max are same
        max_val, min_val = arr[i] , arr[i]
    elif i==j-1: #2 elements
        if arr[j]> arr[i]:
            max_val, min_val = arr[j], arr[i]
        else: 
            max_val, min_val = arr[i], arr[j]
    else:
        mid = i + (j-i)//2    #Divide
        
        # recursion --> conqure
        max_val1, min_val1 = findMaxandMin(arr, i, mid)
        max_val2, min_val2 = findMaxandMin(arr, mid+1, j)
        
        #combine
        if (max_val1 > max_val2):
            max_val = max_val1
        else:
            max_val = max_val2
        
        if (min_val1 < min_val2):
            min_val = min_val1
        else:
            min_val = min_val2        

    return max_val, min_val

## Driver code
arr = [10, 70, 45, 16, 29, 30, 35, 20, 100000, -2]
i = 0
j = len(arr) - 1
## function calling
max_val, min_val = findMaxandMin(arr, i, j)
print("Maximum and Minimum element in an array is:", max_val, min_val)
```
