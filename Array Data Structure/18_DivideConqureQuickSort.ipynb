```python
## function definition of quickSort
def quickSort(arr, p, q):
    if p < q:
        mid = partition(arr, p, q) ## function calling for partition algorithm

        quickSort(arr, p, mid-1) ## recursive call for left subtree  
        quickSort(arr, mid+1, q) ## recursive call for right subtree
    return arr

def partition(arr, p, q):
    i = p
    pivot = arr[p] ## starting element as a pivot element
    for j in range(i+1, q+1):
        if arr[j] <= pivot:
            i += 1          
            arr[i], arr[j] = arr[j], arr[i]  ## swap between the arr[i] and arr[j]
    arr[i], arr[p] = arr[p], arr[i]        ## swap between the arr[i] and the pivot element
    return i
## Driver code
arr = [20, 10, 5, 70, 50, 89, 34]
p = 0
q = len(arr) - 1

result = quickSort(arr, p, q)
print("Sorted array after applying the quickSort is:", result)


```
