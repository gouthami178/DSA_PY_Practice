```python
import numpy as np
''' Given Row wise sorted array
first interger of each row > last integer of previous row
Given target; if target is present return true else false'''
def searchSortedMatrix(matrix, target):
    [m,n] = np.shape(matrix)
    if m == 0 or n==0:
        return False
    
    left, right = 0 , m*n-1
    while(left <= right):
        mid = left + (right-left)//2
        mid_ele = matrix[mid//n][mid%n]
        if(target == mid_ele):
            return True
        elif target > mid_ele :
            left = mid + 1
        else:
            right = mid-1
    return False

## Driver code
matrix = [[1,3,5,7], [10,11,16,20],[23,30,34,60]]

result = searchSortedMatrix(matrix,11)
print(result)
```
