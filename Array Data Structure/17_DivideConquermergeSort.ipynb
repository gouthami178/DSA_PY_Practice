```python
###implemnet merge sort
def mergeSort(arr, i, j):
    if i < j:
        mid = i + (j-i)//2    # Divide
        mergeSort(arr,i,mid)  ## recursive call -> left subtree
        mergeSort(arr,mid+1,j)## recursive call -> right subtree
        mergeProcedure(arr,i,mid,j) #combine
    return arr

def mergeProcedure(arr,i,mid,j):
    n1 = mid -i+1   ## n1 -> number of elements in the left subarray(mid+1, j)
    n2 = j - mid    ## n2 -> number of elements in the right subarray(mid+1, j)

    leftSubarray = [0] * n1
    rightSubarray = [0] * n2
    for m in range(n1):
        leftSubarray[m] = arr[i+m]
    for n in range(n2):
        rightSubarray[n] = arr[mid+n+1]
    p = q = 0
    k = i
    while p < n1 and q < n2:  ## return sorted array 
        if leftSubarray[p] <= rightSubarray[q]:
            arr[k] = leftSubarray[p]
            p += 1
        else:
            arr[k] = rightSubarray[q]
            q += 1
        k += 1

    while p < n1: ## copy the entire elements from the left subarray
        arr[k] = leftSubarray[p]
        p += 1
        k += 1
    
    while q < n2: ## copy the entire elements from the right subarray
        arr[k] = rightSubarray[q]
        q += 1
        k += 1
         
## Driver code
arr = [50, 70, 65, 13, 80, 62, 98, 27]
i = 0
j = len(arr) - 1
## function calling
result = mergeSort(arr, i, j)
print("Sorted array after applying the merge sort is:", result)
```
